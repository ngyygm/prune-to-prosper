```python
import mteb
```

```python
task_categories = {
    "Classification": [
        'AmazonCounterfactualClassification',
        'AmazonReviewsClassification', 'Banking77Classification',
        'EmotionClassification', 'ImdbClassification',
        'MTOPDomainClassification', 'MTOPIntentClassification',
        'MassiveIntentClassification', 'MassiveScenarioClassification',
        'ToxicConversationsClassification', 'TweetSentimentExtractionClassification'
    ],
    "Clustering": [
        'BiorxivClusteringS2S',
        'MedrxivClusteringS2S',
        'TwentyNewsgroupsClustering'
    ],
    "PairClassification": [
        'SprintDuplicateQuestions', 'TwitterSemEval2015', 'TwitterURLCorpus'
    ],
    "Reranking": [
        'AskUbuntuDupQuestions', 'SciDocsRR', 'StackOverflowDupQuestions'
    ],
    "STS": [
        'BIOSSES', 'SICK-R', 'STS12', 'STS13', 'STS14', 'STS15',
        'STS16', 'STS17', 'STSBenchmark'
    ],
    "Summarization": [
        'SummEval'
    ],
    "Retrieval": [
        'ArguAna', 'CQADupstackEnglishRetrieval',
        'NFCorpus', 'SCIDOCS', 'SciFact'
    ],
}
```

```python
download_dataset_dict = {}
for category, datasets in task_categories.items():
    print(f"{category}:")
    for dataset in datasets:
        tasks = mteb.get_tasks(tasks=[dataset])
        task = tasks[0]
        download_dataset_dict[dataset] = {
            "type": task.metadata.type,
            "path": task.metadata.dataset["path"],
            "revision": task.metadata.revision,
            "split": task.metadata.eval_splits
        }
        print(f"  - {dataset} 描述: {task.metadata.description}")
        print(f"      数据集信息: {task.metadata.dataset}")
        print(f"      任务类型: {task.metadata.type}")
        print(f"      数据模态: {task.metadata.modalities}")
        print(f"      评估划分: {task.metadata.eval_splits}")
        print(f"      语言类型: {task.metadata.eval_langs}")
```

```python
from datasets import load_dataset
import os
import json
data_cache = "./data/hf_cache"
output_dir = "./output/"
os.makedirs(output_dir, exist_ok=True)
for dataset_name, meta in download_dataset_dict.items():
    print(dataset_name)
    print(meta['path'])
    print(meta['type'])
    print(meta['split'])
    print("-----")

    try:
        dataset = load_dataset(
            meta["path"],
            "en",
            revision=meta["revision"],
            split="test",
            cache_dir=data_cache
        )
    except:
        dataset = load_dataset(
            meta["path"],
            "default",
            revision=meta["revision"],
            split="test",
            cache_dir=data_cache
        )
    # print(dataset)
    output_path = os.path.join(output_dir, f"{dataset_name}.jsonl")
    with open(output_path, "w", encoding="utf-8") as f:
        for ex in dataset:
            record = {"type": meta["type"]}
            record.update(ex)
            f.write(json.dumps(record, ensure_ascii=False) + "\n")
```
